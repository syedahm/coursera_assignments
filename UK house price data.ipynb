```python
import pandas as pd
```

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
prices_2020_url = 'http://prod.publicdata.landregistry.gov.uk.s3-website-eu-west-1.amazonaws.com/pp-2020.csv'
```

```python
prices_2020 = pd.read_csv(prices_2020_url, header=None)
```

```python
prices_2020.head()
```

```python
split_postcodes = prices_2020[3].str.split(expand=True)
prices_2020['Postcode district'] = split_postcodes[0]
prices_2020['Postcode end'] = split_postcodes[1]
```

```python
prices_2020[2] = pd.to_datetime(prices_2020[2])
```

```python
london_prices = prices_2020[prices_2020[11]=='LONDON']
```

```python
london_prices.head()
```

```python
avg_prices = london_prices[['Postcode district',1]].groupby('Postcode district').mean().reset_index()
avg_prices[1] = avg_prices[1].astype(int)
```
